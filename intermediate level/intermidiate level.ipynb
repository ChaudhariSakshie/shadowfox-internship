
# Sentiment Analysis of X (Twitter) Data

## Objective
To analyze public sentiment expressed on X (Twitter) by classifying tweets into **Positive, Neutral, and Negative** categories using Natural Language Processing (NLP) techniques, and to identify overall sentiment patterns and insights.

---



## Research Questions
1. What is the overall sentiment distribution on X regarding the discussed topic?
2. Which sentiment (positive, neutral, or negative) dominates public opinion?
3. What insights can be derived from sentiment proportions?


## Import Required Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load Dataset

```python

# Load the dataset
df = pd.read_csv(r"C:\Users\Sakshi\OneDrive\Desktop\SHADOWFOX\TASK2\X data.csv")

# Display first few rows
df.head()
```

## Dataset Information

```python

df.info()
```

## Sentiment Label Mapping

```python

# Mapping sentiment categories
sentiment_map = {
    1: "Positive",
    0: "Neutral",
    -1: "Negative"
}

df["sentiment"] = df["category"].map(sentiment_map)

df.head()
```

## Statistical Analysis of Sentiment

```python

# Sentiment count
sentiment_counts = df["sentiment"].value_counts()

sentiment_counts
```

## Sentiment Percentage Distribution

```python

sentiment_percentage = (sentiment_counts / len(df)) * 100
sentiment_percentage
```

## Visualization: Sentiment Distribution (Bar Chart)

```python

sentiment_counts.plot(kind='bar')
plt.title("Sentiment Distribution on X")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.show()
```

## Visualization: Sentiment Distribution (Pie Chart)

```python

sentiment_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Public Opinion Distribution on X")
plt.ylabel("")
plt.show()
```


## Key Observations
