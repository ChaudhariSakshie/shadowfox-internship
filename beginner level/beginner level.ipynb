üìå ***Matplotlib Graph Types***


***1. Line Plot***

Description: Displays trends over time or continuous data
Use Case: Growth analysis, performance tracking

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.show()
```

***2. Scatter Plot***

Description: Shows relationship between two variables
Use Case: Correlation analysis

```python
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.show()
```

***3. Bar Chart***

Description: Compares categories
Use Case: Sales comparison

```python
categories = ['A', 'B', 'C']
values = [30, 45, 20]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.show()
```

***4. Histogram***

Description: Displays data distribution
Use Case: Frequency analysis

```python
data = [10, 20, 20, 30, 40, 40, 40, 50]

plt.hist(data, bins=5)
plt.title("Histogram Example")
plt.show()
```

***5. Pie Chart***

Description: Shows proportional data
Use Case: Percentage breakdown

```python
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

***üìå Seaborn Graph Types***

***1. Line Plot***

Description: Line plot with statistical confidence intervals
Use Case: Trend visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x=x, y=y)
plt.title("Seaborn Line Plot")
plt.show()
```

***2. Scatter Plot***

Description: Enhanced scatter plot with aesthetics
Use Case: Relationship analysis

```python
sns.scatterplot(x=x, y=y)
plt.title("Seaborn Scatter Plot")
plt.show()
```

***3. Bar Plot***

Description: Displays mean values with confidence intervals
Use Case: Category comparison

```python
sns.barplot(x=categories, y=values)
plt.title("Seaborn Bar Plot")
plt.show()
```

***4. Histogram (Distribution Plot)***

Description: Visualizes data distribution
Use Case: Understanding spread of data

```python
sns.histplot(data, bins=5, kde=True)
plt.title("Seaborn Histogram")
plt.show()
```

***5. Box Plot***

Description: Shows data distribution and outliers
Use Case: Outlier detection

```python
sns.boxplot(data=data)
plt.title("Seaborn Box Plot")
plt.show()
```

***üîç Comparison Between Matplotlib and Seaborn***


üîπ Matplotlib

Low-level visualization library in Python

Provides complete control over plot elements

Requires more lines of code for styling

Suitable for custom and complex static plots

Does not provide built-in statistical plots

Learning curve is slightly higher for beginners

üîπ Seaborn

High-level visualization library built on Matplotlib

Provides attractive default themes and color palettes

Requires fewer lines of code

Best suited for statistical and exploratory data analysis

Supports direct plotting from Pandas DataFrames

Very beginner-friendly

| Feature                | Matplotlib | Seaborn           |
| ---------------------- | ---------- | ----------------- |
| Ease of Use            | Moderate   | Easy              |
| Customization          | Very High  | Medium            |
| Default Styling        | Basic      | Excellent         |
| Statistical Plots      | Limited    | Strong            |
| Interactivity          | No         | No                |
| Large Dataset Handling | Good       | Good              |
| Learning Curve         | Steeper    | Beginner-friendly |
